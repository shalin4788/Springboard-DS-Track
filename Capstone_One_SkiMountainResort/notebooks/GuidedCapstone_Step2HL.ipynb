Let's review each of these columns in the data.


*   fastEight is the number of fast eight-person ski lifts at each of the resorts. In addition, you have the counts of all the other types of chairs at each resort and the total number of chairs as well. You can simply subtract the number of fastEight lifts from the total to determine if any of the missing values can be backfilled or if they should be equal to zero. Follow the next step below and see for yourself how to review and backfill these missing values. These columns contain all the resort lift counts by type.

<font color='teal'> **In order to determine an appropriate fill value for the fastEigth column, you need to calculate the difference between the sum of all the chairlift columns and the `total_chairs` column.**</font>

Calculate the difference between the total_chairs column and sum of all chairlift columns across all rows. Make sure because dtype is converted to float to avoid any mathematical error since columns may have float datatype as you had seen in the dtype determination at an earlier step 

```python
ski_df['total_chairs'].astype(float) - ski_df.loc[(slice(None),slice('trams','surface'))].sum(axis=1)
```

As you can see from running the above cell, there are no **fastEight** lift values missing because the difference between the sum of all the lift columns and the 'total_chairs' column is zero for every row in the dataframe. This confirms that you can therefore replace the missing values for the fastEight rows with zeros.

<font color='teal'> **In the cell below, replace Nan's in 'fastEight' with zero.**</font>

Hint: you could use the `fillna()` method here. 

```python
ski_df['fastEight'].isnull().sum()
```

Replace the fastEight rows that have missing values with zeros

```python
ski_df['fastEight'].fillna(0, inplace = True)
```

```python
ski_df
```

```python
ski_df['fastEight'].isnull().sum()
```

Consider the next column with missing values, '**NightSkiing_ac**'. This column represents the number of acres of night skiing availble at this resort. Not all resorts have night skiing because it requires stadium lights and is not that popular with skiers and snowboarders. Given that information, it is also safe to assume that if there are missing values in this column, it is due to there not being any night skiing acres available at that resort so you can fill missing values with zeros safely. 

```python
ski_df['NightSkiing_ac'].isnull().sum()
```

```python
ski_df['NightSkiing_ac'].fillna(0, inplace = True)
ski_df
```

```python
ski_df['NightSkiing_ac'].isnull().sum()
```

Next, let's consider the ticket price columns '**AdultWeekday**' & '**AdultWeekend**'. You know that there has to be a price for these that happens to be missing so, in this case, you cannot fill the missing values with zeros. Instead, you have a few options about what value you can use to fill the missing rows. You can simply take the mean of the column in consideration or you could backfill or forwardfill from the resort in the row before or after your missing value. Backfill and forward fill would be more reasonable in a situation where the data are in a particular order where the rows near each other are similar to each other in the price column. You could also build an imputation model to use a model to predict the value based on the data values in adjacent columns as well as adjacent rows. 

<font color='teal'> **Run the cell below to replace Nan's in 'AdultWeekday' with the mean value.**</font>

```python
ski_df['AdultWeekday'].fillna((ski_df['AdultWeekday'].mean()), inplace=True)
```

<font color='teal'> **Use what you know about the remaining columns missing data and decide whether to fill with a zero or to fill with a mean.**</font>


```python
ski_df['AdultWeekend'].fillna((ski_df['AdultWeekend'].mean()), inplace=True)
```

Next, let's consider the days open columns '**daysOpenLastYear**' & '**projectedDaysOpen**'. You know that the days opened during a given year OR projected for next year is independent for each resort. There could be missing values for a given resort for either of these columns possibly because the resort is under construction and shut down for the season etc. So if there are missing values for any resorts, it is safe to fill them with 0. However, lets check the 2 columns and # of open days last year and projected for next year by each state to make more sense out of this consensus. After that, fill the missing values with 0

```python
ski_df[['state','daysOpenLastYear','projectedDaysOpen']].sort_values(by = 'state', ascending=False)
```

```python
#ski_df[['daysOpenLastYear','projectedDaysOpen']].fillna(0, inplace = True)
ski_df.update(ski_df[['daysOpenLastYear','projectedDaysOpen']].fillna(0))
ski_df
```

```python
ski_df[['daysOpenLastYear','projectedDaysOpen']].isnull().sum()
```

Furthermore, '**TerrainParks**', '**yearsOpen**' and '**averageSnowfall**' are independent values for each rows. 

1) # of years open can be missing in case a resort was recently opened. There is a small number for it and can be safely set to 0 if NaN found. 

2) Similarly, TerrainParks and can be 0 for recorts depending on geography and types of ski sports played. Not all ski activities happen on terrain, so there is no need to do any other computation on it while dealing with the missing values. 

3) Also, for a given year, it is possible that a resort or region may not have experienced snowfall, so missing values can be replaced by 0.

**Note**: # of missing values in TerrainParks are 51 and # of missing values for Skiable terrains are 3. So, '**SkiableTerrain_ac**' could be missing due to 1) terrainParks not available in few resorts 2) Due to no snowfall that year
While it may be okay to substitute ticket price with mean ticket price, but since 'Skiable Terrain area' is a value specific to a given resort, it may not make sense to apply backfill, front fill or even other mathematical computation like mean etc. that we did for the ticket price to fill up missing values, so it is safe to substitute missing values with 0

```python
ski_df.loc[:,['TerrainParks','SkiableTerrain_ac','averageSnowfall']].isnull().sum()
```

```python
ski_df.update(ski_df[['TerrainParks','yearsOpen','averageSnowfall']].fillna(0))
```

```python
ski_df['SkiableTerrain_ac'].fillna(0, inplace = True)
ski_df
```

**Snow Making_ac** which indicates the average area covered by snow making machines in acre can be backfilled OR front filled after sorting the column by 'state'. It is okay to group by state and take the adjacent values

```python
ski_df['Snow Making_ac'].isnull().sum()
```

```python
ski_df['Snow Making_ac'] = ski_df.groupby('state')['Snow Making_ac'].fillna(method='ffill')
```

```python
ski_df['Snow Making_ac'].isnull().sum()
```

```python
ski_df['Snow Making_ac'].fillna(0, inplace = True)
```

```python
ski_df['Snow Making_ac'].isnull().sum()
```

Since the % of missing values for '**Runs**' is greater than the % of missing values for '**LongestRun_mi**' (See below) which is not technically possible and since there are just 4 missing 'Runs' values across all rows, it is safe to:

1) Replace 'LongestRun_mi' missing values with 0

2) Replace 'Runs' with ffill or bfill method similar to how was done for 'Snow Making_ac'

```python
ski_df.loc[:,['Runs','LongestRun_mi']].isnull().sum()
```

```python
ski_df['LongestRun_mi'].fillna(0, inplace = True)
```

```python
ski_df['Runs'] = ski_df.groupby('state')['Runs'].fillna(method='ffill')
```

```python
ski_df[['Runs','Snow Making_ac']].isnull().sum()
```

```python
ski_df.isnull().sum()
```
